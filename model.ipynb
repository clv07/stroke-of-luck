```python
import csv

results = {}
csv_path = 'data/results.csv'

with open(csv_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        file_name = row['patient_num']
        if row['codes'].strip():
            codes = [code.strip() for code in row['codes'].split(',')]
        else:
            codes = []
        results[file_name] = codes

results
```

```python
df['truths'] = df['PatientID'].map(results)

print(df.head())
```

```python
empty_count_before = df[df['truths'].isnull()]
df = df.dropna(subset=['truths'])
empty_count_after = df[df['truths'].isnull()]

print("Number of empty cells:", empty_count_before)
print("Number of empty cells:", empty_count_after)
```

```python
null_counts = df.isnull().sum()
print(null_counts)
```

```python
missing_columns = null_counts[null_counts > 0]
print(missing_columns)
```

```python
import lightgbm as lgb
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mlb = MultiLabelBinarizer()

# Fit the MultiLabelBinarizer object on the truths column to make a robust list of classes to train on and fit to
mlb.fit(df['truths'])

# make the truths column
y = mlb.transform(df['truths'])

print(len(mlb.classes_))
print(df['truths'])
print("Classes:", mlb.classes_)
print("Binary Matrix:\n", y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['PatientID','Gender', 'PatientAge', 'AcquisitionDateTime_DT', 'truths']), y, test_size=0.2, random_state=42)


lgb_params = {
    'objective': 'binary',
    'boosting_type': 'gbdt',
    'n_estimators': 100,
    'learning_rate': 0.03,
    'random_state': 42,
    'verbose': -1,
}

model = MultiOutputClassifier(lgb.LGBMClassifier(**lgb_params))

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

score = f1_score(y_test, y_pred, average='micro')
print(score)

import joblib
joblib.dump(model, 'model.pkl')


```

```python
from sklearn.metrics import confusion_matrix

n_labels = y_test.shape[1]

for i in range(n_labels):
    cm = confusion_matrix(y_test[:, i], y_pred[:, i])
    TN, FP, FN, TP = cm.ravel()
    fpr = FP / (FP + TN) if (FP + TN) > 0 else 0
    tpr = TP / (TP + FN) if (TP + FN) > 0 else 0
    print(f"Label {i}:")
    print("Confusion Matrix:")
    print(cm)
    print(f"FPR: {fpr:.3f}, TPR (Recall): {tpr:.3f}\n")
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
learning_rates = [0.04,0.035, 0.03, 0.025, 0.02]
scores = {}
for learning_rate in learning_rates:
    lgb_params = {
        'objective': 'binary',
        'boosting_type': 'gbdt',
        'n_estimators': 100,
        'learning_rate': learning_rate,
        'random_state': 42,
        'verbose': -1,
        }
    model = MultiOutputClassifier(lgb.LGBMClassifier(**lgb_params))
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score = f1_score(y_test, y_pred, average='micro')
    scores[(learning_rate)] = score
    print(f"Learning rate: {learning_rate}, F1 score: {score:.4f}")

```
